```python
from nba_api.stats.endpoints import leaguegamefinder
```

```python
games = leaguegamefinder.LeagueGameFinder(
    season_nullable="2024-25"
)

df_2024_season_raw = games.get_data_frames()[0]

df_2024_season_raw.to_csv('nba_league_games_2024-25_raw.csv')
```

# MATCHUP

either 'vs' or '@' in the middle indicating whether home and away team.

team with '@' means away team

team with 'vs' means home

```python
df.head()
```

## Add column HOME_TEAM which will be boolean

game at home will be marked with a true

```python
def isHomeGame(val) -> bool:
    return "vs" in val
```

```python
df["HOME_TEAM"] = df['MATCHUP'].apply(isHomeGame)
```

```python
home_games = df.loc[df["HOME_TEAM"]]
```
