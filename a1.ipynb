# 2024 df Transformation function

```python
import pandas as pd

def isHomeGame(val) -> bool:
    return "vs" in val

def transform_games_df(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    df["GAME_ID"] = df["GAME_ID"].astype(int)
    df["GAME_DATE"] = pd.to_datetime(df["GAME_DATE"], errors="raise")
    df["HOME_TEAM"] = df["MATCHUP"].apply(isHomeGame)

    return df
```

```python
df_final = df_2024_season_raw
```

```python
df_final_standard = transform_games_df(df_final)
```

```python
df_final_standard.info()
```

```python
df_final_standard.to_csv("nba_league_games_2024-25_standard.csv", index=False)
```
